```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate 100 dates in a series
dates = pd.date_range(start="2023-01-01", periods=100, freq="D")

# Generate 100 random numbers between 1 and 10
prices = np.random.randint(1, 11, size=100)

# Create a DataFrame
df = pd.DataFrame({"Date": dates, "Price": prices})

# Save the DataFrame to an Excel file
df.to_excel("EXCEL SHEET.xlsx", index=False)

# Plot the chart
plt.figure(figsize=(11, 7))
plt.plot(df["Date"], df["Price"], label="Price")
plt.plot(df["Date"], df["Price"].rolling(window=10).mean(), label="10-day MA")
plt.plot(df["Date"], df["Price"].rolling(window=50).mean(), label="50-day MA")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.title("Price")

# Save the plot to an image file
plt.savefig("Picture2.png")

# Show the plot
plt.show()

#links of Excel file and the chart image on a cloud platform
excel_link = "https://assignmentfor.s3.ap-south-1.amazonaws.com/EXCEL%20SHEET.xlsx?response-content-disposition=inline&X-Amz-Security-Token=IQoJb3JpZ2luX2VjEDgaCmFwLXNvdXRoLTEiRzBFAiEAzJe5%2B03NuW0QL60cyh3nrlN%2BCLKPMKyxWcMkjrC916wCICt9hq2lHXSecTJ2pXZm0VxUMq%2BO"
chart_link = "https://assignmentfor.s3.ap-south-1.amazonaws.com/Picture2.png?response-content-disposition=inline&X-Amz-Security-Token=IQoJb3JpZ2luX2VjEDgaCmFwLXNvdXRoLTEiRzBFAiEAzJe5%2B03NuW0QL60cyh3nrlN%2BCLKPMKyxWcMkjrC916wCICt9hq2lHXSecTJ2pXZm0VxUMq%2BO5cRjZCJN%2BCp%2FvUDHKu0CCKH%2F%2F%2F%2F%2F%2F%2F%2F%2F%2FwEQABoMMjM0NzAxNzY3MDcwIgx7%2B6vhL7jzDxTeXEgqwQJyKEauSf6H19gTskaCtwoQbCOOso%2F1LUn42dN1AEeM%2BoJbr92pQ38H9EWIfP9vJ4ThC1dF1Zf7J82FUjc6M2GODqtDgJ6LCopU3aZSqrIwn7q33xTj%2BycGC8RQL4fpMoEiel0AjWGu%2B%2FJ5vAvDoVKg9RboqC7ks3jpROqAVWytRatiUIf7HSqC3dpj9OVVslFNs1ZHCfROMFhETrNhr%2B8edvAIFLUWVhRqgryzD%2Bm1sTh70YBqbgNj4Mj51kZkVUihTsdvD2sfFKTmEwIl%2BvbSW2FvNyQASVKMbIUjaD6JqbIzz%2BltcbK6GvMvOwnSm8etsOxmvpmSEZ8MENwpysY7kqLHDzfXXqlAdFr7fa0pKbDeqBUWfxSrY2k0eiE%2BSToWE2F63siC1P8LpWnHH%2F9G4Rt7VzY7TO7mbhwaSS1BGjEwy%2BXFqwY6swKeTX5eEvbnRS5AQJhSMZBMW9dsAibNaX9UqL%2Fmo5bnNe0VavSFiJkYJ2v3aLRfdACJcjhoZ8ynXiwqf%2BSWk56otpEEXCx9Cg1HuCx1nrB4MTPrWp0yXjz7AfBakq%2BhWOyU2gCa08ZMk6BKmkmpzAQwnzx%2FaQS8RQANraxaBrSvblP8MmUX6ga9VezLeAdbp4dktjtBfAjUre%2BXR3XZ8kxq0%2B1tQlCRHKuj4ZCqFUur5WLd4oPe8INbVFE6Zt0vAvccJBfbjtee14e3GWyyV2PqLTH3tbdsv3RN7S39VhCL1uiZuT%2BgkcanSxw53YvjNnXfmGwkPfdA9cPRAgwNx7hUx1SjwM6gZfwPmfYiBEt1ZOofPYT5KjAeSQFTeifjpJutaj83SgnDIHCtOlP8ggGW6Aor&X-Amz-Algorithm=AWS4-HMAC-SHA256&X-Amz-Date=20231207T074753Z&X-Amz-SignedHeaders=host&X-Amz-Expires=43200&X-Amz-Credential=ASIATNJKQYGPHGWCLEE5%2F20231207%2Fap-south-1%2Fs3%2Faws4_request&X-Amz-Signature=e927cf7ad576b26f9de25f7732a2f7f3cd66743b9ed7ef3f2a6d2aae1eb09cc2"

# Printing the link so you can directed towards the hosted site
print(f"Excel file: {excel_link}")
print(f"Chart image: {chart_link}")
```
